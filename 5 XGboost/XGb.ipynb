```python
import numpy as np
import pandas as pd


```

```python
train_set=pd.read_csv('train_set.csv')
test_set=pd.read_csv('test_set.csv')
```

```python
train_set.head(2)
```

```python
train_set.describe()
```

```python
train_set=train_set.drop('fnlwgt', axis=1)
train_set['wage_class']=train_set['wage_class'].replace([' <=50K', ' >50K'],[0,1])
train_set.head(2)
```

```python
train_set.head(1)
```

```python
y=train_set[['wage_class']]
x=train_set.drop('wage_class', axis=1)
x.head(1)
```

```python
data_count={"col":[],'unique':[], "dtype":[]}
for i in x.columns:
    data_count['col'].append(i)
    data_count['unique'].append(len(train_set[i].unique()))
    data_count['dtype'].append(train_set[i].dtype)

data_count=pd.DataFrame(data_count)
data_count
```

```python
int_col=np.array(data_count[data_count['dtype']=="int64"]['col'])
ord_col=np.array(data_count[data_count['dtype']=="object"]['col'])
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
import xgboost as xgb
ord_transformer=ColumnTransformer(
    transformers=[
        ("onehotencode", OneHotEncoder(drop='first'), ord_col)  
    ])


clf=Pipeline([
    ("ord_pre", ord_transformer),
    ('clf', xgb.XGBClassifier())
    ])
clf
```

```python
clf.fit(x,y)
```

```python
test_set=test_set.drop('fnlwgt', axis=1)
test_set['wage_class']=test_set['wage_class'].replace([' <=50K.', ' >50K.'],[0,1])
test_set.head(2)
```

```python
x_test=test_set.drop('wage_class', axis=1)
y_test=test_set[['wage_class']]
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
